```python
#IMPORTANT
#When pulling, please download the data and upload to your notebook: 
#1. "ipc_afg_area_long.csv" from https://data.humdata.org/dataset/afghanistan-acute-food-insecurity-country-data

#IMPORTANT: When pulling, please change MY EMAIL to YOUR email address. Example below: 
#pd.read_csv("dbfs:/Workspace/Users/<YOUR EMAIL>/")

import pandas as pd
import numpy as np

#Import data for Afghanistan's food insecurity
afghanistan_food_insecurity = pd.read_csv("/Workspace/Users/elizabethnguyen@ufl.edu/ipc_afg_area_long.csv")

#Exclude urban populations right now...maybe come back and combine rural+urban
afghanistan_food_insecurity = afghanistan_food_insecurity[afghanistan_food_insecurity["Area"].str.contains("Urban") == False]

#Drop the first and second projections (these data points are the UN predicting the future)
afghanistan_food_insecurity = afghanistan_food_insecurity[afghanistan_food_insecurity["Validity period"] == "current"]

#Drop "To" and "Date of analysis" columns
afghanistan_food_insecurity = afghanistan_food_insecurity.drop(columns = ["To"])
afghanistan_food_insecurity = afghanistan_food_insecurity.drop(columns = ["Date of analysis"])

#Only data from 2025
afghanistan_food_insecurity['From'] = pd.to_datetime(afghanistan_food_insecurity['From'])
afghanistan_food_insecurity = afghanistan_food_insecurity[afghanistan_food_insecurity["From"].dt.year == 2025]
#Might change from standard datetime data to get rid of time (00:00.000)

#Only phases 3, 4, and 5 (doesn't include phase all and 3+) 
#These phases are what the UN identifies as "crisis"
afghanistan_food_insecurity = afghanistan_food_insecurity[
    (afghanistan_food_insecurity["Phase"] == "3") | 
    (afghanistan_food_insecurity["Phase"] == "4") | 
    (afghanistan_food_insecurity["Phase"] == "5")]

#Displays the whole table
display(afghanistan_food_insecurity)
```
